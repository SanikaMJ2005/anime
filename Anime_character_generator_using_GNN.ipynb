```python
import torch
import matplotlib.pyplot as plt
from diffusers import StableDiffusionPipeline

pipe = StableDiffusionPipeline.from_pretrained(
    "hakurei/waifu-diffusion",
    torch_dtype=torch.float16,
    safety_checker=None   # disable false NSFW blocking
).to("cuda")

prompt = (
    "anime character portrait, clean face, fully clothed, "
    "simple background, natural colors, soft lighting, "
    "cute anime style, high quality, sharp details"
)

negative_prompt = (
    "nsfw, nude, cleavage, erotic, sexy, sexualized, "
    "distorted face, extra eyes, abstract, blurry"
)

image = pipe(
    prompt,
    negative_prompt=negative_prompt,
    num_inference_steps=40,
    guidance_scale=8
).images[0]

plt.imshow(image)
plt.axis("off")
plt.show()

image.save("safe_anime_output.png")
print("✅ Safe anime image generated and saved as safe_anime_output.png")
```

```python
import torch
import matplotlib.pyplot as plt
from diffusers import StableDiffusionPipeline

pipe = StableDiffusionPipeline.from_pretrained(
    "hakurei/waifu-diffusion",
    torch_dtype=torch.float16,
    safety_checker=None
).to("cuda")

generator = torch.Generator("cuda").manual_seed(1234)  # ← change this number

prompt = (
    "anime character portrait, clean face, fully clothed, "
    "simple background, natural colors, soft lighting, "
    "cute anime style, high quality"
)

negative_prompt = (
    "nsfw, nude, cleavage, erotic, distorted face, "
    "extra eyes, blurry, abstract"
)

image = pipe(
    prompt,
    negative_prompt=negative_prompt,
    num_inference_steps=40,
    guidance_scale=8,
    generator=generator
).images[0]

plt.imshow(image)
plt.axis("off")
plt.show()
```
